```python
import speech_recognition as sr
import cv2
import getpass

class VoiceScanner:
    def recognize_voice(self):
        recognizer = sr.Recognizer()
        microphone = sr.Microphone()

        with microphone as source:
            print("Speak now...")
            recognizer.adjust_for_ambient_noise(source)
            audio = recognizer.listen(source)

        try:
            recognized_text = recognizer.recognize_google(audio)
            print("Recognized voice: ", recognized_text)
            return recognized_text
        except sr.UnknownValueError:
            print("Unable to recognize voice.")
            return None

class EyeScanner:
    def recognize_eye(self):
        # Load the pre-trained Haar cascade XML file for eye detection
        eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')

        # Open the default camera
        cap = cv2.VideoCapture(0)

        while True:
            # Read the current frame
            ret, frame = cap.read()

            # Convert the frame to grayscale
            gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

            # Detect eyes in the grayscale frame
            eyes = eye_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

            # Draw rectangles around the detected eyes
            for (x, y, w, h) in eyes:
                cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)

            # Display the frame
            cv2.imshow('Eye Scan', frame)

            # Exit the loop when 'q' is pressed
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break

        # Release the video capture and close the OpenCV windows
        cap.release()
        cv2.destroyAllWindows()

class SecuritySystem:
    def __init__(self):
        self.voice_scanner = VoiceScanner()
        self.eye_scanner = EyeScanner()
        self.authorized_users = {
            'user1': {
                'voice': 'voice1',
                'eye': 'eye1'
            },
            'user2': {
                'voice': 'voice2',
                'eye': 'eye2'
            }
        }

    def authenticate(self, username):
        if username in self.authorized_users:
            voice_sample = self.voice_scanner.recognize_voice()
            eye_scan = self.eye_scanner.recognize_eye()

            if voice_sample == self.authorized_users[username]['voice'] and \
                    eye_scan is not None:
                print("Access granted!")
            else:
                print("Access denied!")
        else:
            print("User not found!")

def main():
    security_system = SecuritySystem()
    username = input("Enter your username: ")
    password = getpass.getpass("Enter your password: ")

    # Authenticating using voice and eye scan
    if password == 'secret':
        security_system.authenticate(username)
    else:
        print("Incorrect password!")

if __name__ == "__main__":
    main()
```
